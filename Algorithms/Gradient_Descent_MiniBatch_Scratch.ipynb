# **Linear Regression**

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_diabetes
X, y = load_diabetes(return_X_y=True)

print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

# **My Class**

```python
import random

class MyGD_MiniBatch:
  def __init__(self, batchSize, lRate=0.01, epochs=100):
    self.coef_ = None
    self.intercept_ = None
    self.batchSize = batchSize
    self.lRate = lRate
    self.epochs = epochs

  def fit(self, X_train, y_train):
    self.intercept_ = 0
    self.coef_ = np.ones(X_train.shape[1])

    for i in range(self.epochs):
      for j in range(int(X_train.shape[0] / self.batchSize)):
        idx = random.sample(range(X_train.shape[0]), self.batchSize)

        y_hat = np.dot(X_train[idx], self.coef_) + self.intercept_

        intercept_der = -2 * np.mean(y_train[idx] - y_hat)
        self.intercept_ = self.intercept_ - (self.lRate * intercept_der)

        coef_der = -2 * np.dot((y_train[idx] - y_hat), X_train[idx])
        self.coef_ = self.coef_ - (self.lRate * coef_der)

    print(self.coef_)
    print(self.intercept_)

  def predict(self, X_test):
    return np.dot(X_test, self.coef_) + self.intercept_
```

```python
gd = MyGD_MiniBatch(batchSize=45, epochs=100, lRate=0.1)

import time
start = time.time()

gd.fit(X_train, y_train)

print(time.time() - start)
```

```python
y_pred = gd.predict(X_test)
r2_score(y_test, y_pred)
```

# **SGD Regressor**

```python
from sklearn.linear_model import SGDRegressor
sgd = SGDRegressor(learning_rate='constant', eta0=0.1)
```

```python
batchSize = 45

for i in range(100):
  idx = random.sample(range(X_train.shape[0]), batchSize)
  sgd.partial_fit(X_train[idx], y_train[idx])
```

```python
print(sgd.coef_)
print(sgd.intercept_)
```

```python
y_pred = sgd.predict(X_test)
r2_score(y_test, y_pred)
```
