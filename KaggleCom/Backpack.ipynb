```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy='most_frequent')
```

```python
df['Size'] = si.fit_transform(df[['Size']])[:, 0]
df_test['Size'] = si.transform(df_test[['Size']])[:, 0]
df['Laptop Compartment'] = si.fit_transform(df[['Laptop Compartment']])[:, 0]
df_test['Laptop Compartment'] = si.transform(df_test[['Laptop Compartment']])[:, 0]
df['Waterproof'] = si.fit_transform(df[['Waterproof']])[:, 0]
df_test['Waterproof'] = si.transform(df_test[['Waterproof']])[:, 0]
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Small','Medium','Large'],['No','Yes'], ['No','Yes']])
```

```python
df[['Size', 'Laptop Compartment', 'Waterproof']] = oe.fit_transform(df[['Size', 'Laptop Compartment', 'Waterproof']])
df_test[['Size', 'Laptop Compartment', 'Waterproof']] = oe.transform(df_test[['Size', 'Laptop Compartment', 'Waterproof']])
```

```python
df.head()
```

```python
df['Brand'] = si.fit_transform(df[['Brand']])[:, 0]
df_test['Brand'] = si.transform(df_test[['Brand']])[:, 0]
df['Material'] = si.fit_transform(df[['Material']])[:, 0]
df_test['Material'] = si.transform(df_test[['Material']])[:, 0]
df['Style'] = si.fit_transform(df[['Style']])[:, 0]
df_test['Style'] = si.transform(df_test[['Style']])[:, 0]
df['Color'] = si.fit_transform(df[['Color']])[:, 0]
df_test['Color'] = si.transform(df_test[['Color']])[:, 0]
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')
```

```python
df_ohe = ohe.fit_transform(df[['Brand', 'Material', 'Style', 'Color']])
df_test_ohe = ohe.transform(df_test[['Brand', 'Material', 'Style', 'Color']])
```

```python
df_ohe = pd.DataFrame(df_ohe, columns=ohe.get_feature_names_out(['Brand', 'Material', 'Style', 'Color']))
df_test_ohe = pd.DataFrame(df_test_ohe, columns=ohe.get_feature_names_out(['Brand', 'Material', 'Style', 'Color']))
```

```python
df = df.drop(columns=['Brand', 'Material', 'Style', 'Color'])
df_test = df_test.drop(columns=['Brand', 'Material', 'Style', 'Color'])
```

```python
df = pd.concat([df, df_ohe], axis=1)
df_test = pd.concat([df_test, df_test_ohe], axis=1)
```

```python
df.head()
```

```python
si_w = SimpleImputer()
df['Weight Capacity (kg)'] = si_w.fit_transform(df[['Weight Capacity (kg)']])[:, 0]
df_test['Weight Capacity (kg)'] = si_w.transform(df_test[['Weight Capacity (kg)']])[:, 0]
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['id'], inplace=True)

id = df_test['id']
df_test.drop(columns=['id'], inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
```

```python
df[['Weight Capacity (kg)']] = mms.fit_transform(df[['Weight Capacity (kg)']])
df_test[['Weight Capacity (kg)']] = mms.transform(df_test[['Weight Capacity (kg)']])
```

```python
df.corr()['Price']
```

```python
X_train = df.drop('Price', axis=1)
y_train = df['Price']

X_test = df_test
y_test = pd.read_csv('sample_submission.csv')['Price']
```

```python
from xgboost import XGBRegressor

model = XGBRegressor(
    objective='reg:squarederror',
    learning_rate=0.04,
    max_depth=4,
    n_estimators=200,
    subsample=0.8,
    random_state=2
)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import root_mean_squared_error
score = root_mean_squared_error(y_test, y_pred)
score
```
