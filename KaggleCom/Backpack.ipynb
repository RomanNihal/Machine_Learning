```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
df_1 = pd.read_csv('training_extra.csv')
```

```python
df = pd.concat([df, df_1], ignore_index=True)
df_test = pd.read_csv('test.csv')
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df_test.isnull().sum()
```

```python
df['Laptop Compartment'].unique()
```

```python
df['Laptop Compartment'] = df['Laptop Compartment'].map({'Yes': 1, 'No': 0}).fillna(0.5)
df['Waterproof'] = df['Waterproof'].map({'Yes': 1, 'No': 0}).fillna(0.5)
```

```python
df_test['Laptop Compartment'] = df_test['Laptop Compartment'].map({'Yes': 1, 'No': 0}).fillna(0.5)
df_test['Waterproof'] = df_test['Waterproof'].map({'Yes': 1, 'No': 0}).fillna(0.5)
```

```python
size_mapping = {'Small': 1, 'Medium': 2, 'Large': 3}
df['Size'] = df['Size'].map(size_mapping)
df_test['Size'] = df_test['Size'].map(size_mapping)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')
df['Size'] = imputer.fit_transform(df[['Size']])[:, 0]
df_test['Size'] = imputer.fit_transform(df_test[['Size']])[:, 0]
```

#Label Encoder

```python
from sklearn.preprocessing import LabelEncoder

features = ['Brand', 'Material', 'Style', 'Color']

for feature in features:
    le = LabelEncoder()
    df[feature] = le.fit_transform(df[feature])

    df_test[feature] = le.fit_transform(df_test[feature])
```

```python
df['Weight Capacity (kg)'] = df['Weight Capacity (kg)'].fillna(df['Weight Capacity (kg)'].mean())
df_test['Weight Capacity (kg)'] = df_test['Weight Capacity (kg)'].fillna(df_test['Weight Capacity (kg)'].mean())
```

```python
df.drop(columns=['id', 'Waterproof'], inplace=True)

id = df_test['id']
df_test.drop(columns=['id', 'Waterproof'], inplace=True)
```

```python
X = df.drop('Price', axis=1)
y = df['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
import sklearn.metrics as metrics
```

```python
from xgboost import XGBRegressor

model = XGBRegressor(
    objective='reg:squarederror',
    learning_rate=0.05,
    max_depth=4,
    n_estimators=200,
    subsample=0.8,
    random_state=3
)

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print('RMSE:', metrics.root_mean_squared_error(y_test, y_pred))
```

```python
y_pred = model.predict(df_test)

submission_df = pd.DataFrame({'id': id, 'Price': y_pred})
submission_df.to_csv('submission.csv', index=False)
```
