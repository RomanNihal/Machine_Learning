```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
titanic = pd.read_csv('titanic.csv')
titanic.head()
```

###Numerical - Numerical

```python
sns.scatterplot(x='total_bill', y='tip', data=pd.concat([tips]))
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=pd.concat([tips]))
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', style='smoker', data=pd.concat([tips]))
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', style='smoker', size='size', data=pd.concat([tips]))
```

###Numerical - Categorical

```python
sns.barplot(x='Pclass', y='Age', data=pd.concat([titanic]))
```

```python
sns.barplot(x='Pclass', y='Fare', data=pd.concat([titanic]))
```

```python
sns.barplot(x='Pclass', y='Fare', hue='Sex', data=pd.concat([titanic]))
```

```python
sns.boxplot(x='Sex', y='Age', data=pd.concat([titanic]))
```

```python
sns.boxplot(x='Sex', y='Age', hue='Survived', data=pd.concat([titanic]))
```

```python
sns.histplot(
    data=titanic, x='Age', kde=True,
    stat='density', kde_kws=dict(cut=3),
    alpha=.2, edgecolor=(1, 1, 1, 1),
    hue='Survived'
)
```

###Categorical - Categorical

```python
pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
titanic.groupby('Pclass').mean(numeric_only=True)['Survived']*100
```

```python
(titanic.groupby('Pclass').mean(numeric_only=True)['Survived']*100).plot(kind='bar')
```

```python
(titanic.groupby('Sex').mean(numeric_only=True)['Survived']*100).plot(kind='bar')
```

```python
(titanic.groupby('Embarked').mean(numeric_only=True)['Survived']*100).plot(kind='bar')
```

```python
(titanic.groupby('Embarked').mean(numeric_only=True)['Survived']*100).plot(kind='bar')
```

```python
pd.crosstab(titanic['Parch'], titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
```

```python
sns.pairplot(iris, hue='species')
```
