```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
df.head()
```

```python
df = df.iloc[:, 2:]
```

```python
df.head()
```

```python
X = df.drop('Purchased', axis=1)
y = df['Purchased']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_s = sc.fit_transform(X_train)
X_test_s = sc.transform(X_test)
```

```python
X_train.describe()
```

```python
X_train_s = pd.DataFrame(X_train_s, columns=X_test.columns)
X_test_s = pd.DataFrame(X_test_s, columns=X_test.columns)
```

```python
X_train_s.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.scatterplot(x='Age', y='EstimatedSalary', data=pd.concat([X_train]), ax=axes[0])
axes[0].set_title('Original Data')

sns.scatterplot(x='Age', y='EstimatedSalary', data=pd.concat([X_train_s]), ax=axes[1])
axes[1].set_title('Scaled Data')

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.kdeplot(x='Age', data=X_train, ax=axes[0], fill=True)
axes[0].set_title('KDE of Original Data')
sns.kdeplot(x='EstimatedSalary', data=X_train, ax=axes[0],  color = 'red', fill = True)

sns.kdeplot(x='Age', data=X_train_s, ax=axes[1], fill = True)
axes[1].set_title('KDE of Scaled Data')
sns.kdeplot(x='EstimatedSalary', data=X_train_s, ax=axes[1], color = 'red', fill=True)

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.kdeplot(x='Age', data=X_train, ax=axes[0], fill=True)
axes[0].set_title('Distribution of Original Age')

sns.kdeplot(x='Age', data=X_train_s, ax=axes[1], fill = True)
axes[1].set_title('Distribution of Scaled Age')

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.kdeplot(x='EstimatedSalary', data=X_train, ax=axes[0], fill=True)
axes[0].set_title('Distribution of Original EstimatedSalary')

sns.kdeplot(x='EstimatedSalary', data=X_train_s, ax=axes[1], fill = True)
axes[1].set_title('Distribution of Scaled EstimatedSalary')

plt.show()
```
