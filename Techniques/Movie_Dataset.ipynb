```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv('credits.csv')
credits.info()
```

```python
keywords = pd.read_csv('keywords.csv')
keywords.info()
```

```python
merged_df = pd.merge(credits, keywords, on='id')
merged_df.info()
```

## ValueError: invalid literal for int() with base 10: '1997-08-20'

## Solution:
#### movies_metadata = pd.read_csv('movies_metadata.csv')
#### movies_metadata['id'] = pd.to_numeric(movies_metadata['id'], errors='coerce')
#### movies_metadata['id'] = movies_metadata['id'].astype(pd.Int64Dtype())
#### movies_metadata.info()

```python
movies_metadata = pd.read_csv('movies_metadata.csv')

values_to_remove = ['1997-08-20', '2012-09-29', '2014-01-01']
movies_metadata = movies_metadata[~movies_metadata['id'].isin(values_to_remove)]

movies_metadata = movies_metadata.astype({"id": np.int64})
movies_metadata.info()
```

```python
final_df = pd.merge(merged_df, movies_metadata, on='id')
final_df.info()
```

```python
selected_columns = ['id', 'genres', 'original_title', 'poster_path', 'overview',
                    'original_language', 'release_date', 'keywords', 'cast', 'crew']
final_df = final_df[selected_columns]

final_df.info()
```

```python
from typing_extensions import final
final_df = final_df.rename(columns={
    'original_title': 'title',
    'poster_path': 'poster',
    'original_language': 'language',
    'release_date': 'release'
})

final_df.info()
```

```python
final_df.drop_duplicates(inplace=True)
final_df.dropna(inplace=True)
final_df.info()
```

```python
final_df.drop_duplicates(subset=['title'], keep='first', inplace=True)
```

```python
final_df.drop_duplicates(subset=['poster'], keep='first', inplace=True)
```

```python
final_df.drop_duplicates(subset=['overview'], keep='first', inplace=True)
```

```python
final_df = final_df[final_df['genres'] != "[]"]
```

```python
final_df = final_df[final_df['keywords'] != "[]"]
```

```python
final_df = final_df[final_df['cast'] != "[]"]
```

```python
final_df = final_df[final_df['crew'] != "[]"]
```

```python
final_df.info()
```

```python
final_df.to_csv('movies.csv', index=False)
```
