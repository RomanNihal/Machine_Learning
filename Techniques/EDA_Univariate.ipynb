Univariate analysis means analysis of a single column.

There are two types of data:
1. Numerical : Age, salary etc.
2. Categorical: Nationality, Gender etc.

```python
import pandas as pd
```

```python
df=pd.read_csv("titanic.csv")
```

```python
df.head()
```

# Chategorical data

1. Countplot

```python
import seaborn as sns
```

```python
df['Survived'].value_counts()
sns.countplot(x="Survived",data=df)

#df['Survived'].value_counts().plot(kind='bar')
```

```python
df['Pclass'].value_counts()
sns.countplot(x="Pclass",data=df)

#df['Pclass'].value_counts().plot(kind='bar')
```

```python
df['Sex'].value_counts()
sns.countplot(x="Sex",data=df)

#df['Sex'].value_counts().plot(kind='bar')
```

2. Pie Chart

```python
df['Survived'].value_counts().plot(kind='pie', autopct='%.2f')
```

```python
df['Pclass'].value_counts().plot(kind='pie', autopct='%.2f')
```

# Numerical data

1. Histogram

```python
import matplotlib.pyplot as plt
```

Age is a continuous value and histogram convert it into catagory. It gives count in a range.

```python
plt.hist(df['Age'])
```

```python
plt.hist(df['Age'], bins=5)
```

```python
plt.hist(df['Age'], bins=30)
```

2. Distplot

Using this we can tell a random values probability. In this case, the probability of a person is 20 years old is .028%.

It also tells us is there any skewness in the data or not.

```python
sns.distplot(df['Age'])
```

Another way to find skewness

```python
df['Age'].skew()
```
