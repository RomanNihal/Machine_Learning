```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'
```

```python
page  = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
soup
```

```python
soup.find_all('table')[0]
```

```python
soup.find('table', class_='wikitable sortable sticky-header-multi sort-under')
```

```python
table = soup.find_all('table')[0]
```

```python
titles = table.find_all('th')
titles
```

```python
table_titles = [title.text.strip() for title in titles[0:9]]
table_titles = [title.replace('[note 1]', '').strip() if title.endswith('[note 1]') else title for title in table_titles]
table_titles
```

```python
import pandas as pd
df = pd.DataFrame(columns=table_titles)
df
```

```python
column_data = table.find_all('tr')[2:]

all_rows = []

for row_element in column_data:
    row_data = []

    rank = row_element.find('th')
    if rank:
        row_data.append(rank.text.strip().replace('\n', ''))

    remaining_data = row_element.find_all('td')
    for data_element in remaining_data:
        row_data.append(data_element.text.strip().replace('\n', ''))

    df.loc[len(df)] = row_data
```

```python
df.sample(5)
```

```python
df.drop(columns=[col for col in ['State-owned', 'Ref.'] if col in df.columns], inplace=True)
df.head()
```

```python
df.to_csv('largest_companies_by_revenue.csv', index=False)
```
