Label encoder is for the output column, works similar to ordinal encoder.

Ordinal encoder is for the input column.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('customer.csv')
df.head()
```

```python
df['gender'].unique() #Nominal
```

```python
df['review'].unique() #Ordinal
```

```python
df['education'].unique() #Ordinal
```

```python
df['purchased'].unique() #Nominal
```

```python
df = df.iloc[:, 2:]
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:, 0:2], df.iloc[:, -1], test_size=0.2, random_state=2)
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
X_train = oe.fit_transform(X_train)
X_test = oe.transform(X_test)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```
