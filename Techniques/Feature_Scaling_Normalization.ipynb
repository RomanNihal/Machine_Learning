```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('wine_data.csv', header=None, usecols=[0,1,2])
df.columns=['Class label', 'Alcohol', 'Malic acid']
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.kdeplot(x='Alcohol', data=df, ax=axes[0], fill=True)
axes[0].set_title('Distribution of Alcohol')

sns.kdeplot(x='Malic acid', data=df, ax=axes[1], fill = True)
axes[1].set_title('Distribution of Malic acid')

plt.show()
```

```python
sns.scatterplot(x='Alcohol', y='Malic acid', hue='Class label', data=pd.concat([df]))
```

```python
X = df.drop('Class label', axis=1)
y = df['Class label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
X_train_s = sc.fit_transform(X_train)
X_test_s = sc.transform(X_test)
```

```python
X_train.describe()
```

```python
X_train_s = pd.DataFrame(X_train_s, columns=X_test.columns)
X_test_s = pd.DataFrame(X_test_s, columns=X_test.columns)
```

```python
X_train_s.describe()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.scatterplot(x='Alcohol', y='Malic acid', hue='Class label', data=pd.concat([df]), ax=axes[0])
axes[0].set_title('Original Data')

sns.scatterplot(x='Alcohol', y='Malic acid', hue='Class label', data=pd.concat([df]), ax=axes[1])
axes[1].set_title('Scaled Data')

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.kdeplot(x='Alcohol', data=X_train, ax=axes[0], fill=True)
axes[0].set_title('KDE of Original Data')
sns.kdeplot(x='Malic acid', data=X_train, ax=axes[0],  color = 'red', fill = True)

sns.kdeplot(x='Alcohol', data=X_train_s, ax=axes[1], fill = True)
axes[1].set_title('KDE of Scaled Data')
sns.kdeplot(x='Malic acid', data=X_train_s, ax=axes[1], color = 'red', fill=True)

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.kdeplot(x='Alcohol', data=X_train, ax=axes[0], fill=True)
axes[0].set_title('Distribution of Original Alcohol')

sns.kdeplot(x='Alcohol', data=X_train_s, ax=axes[1], fill = True)
axes[1].set_title('Distribution of Scaled Alcohol')

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.kdeplot(x='Malic acid', data=X_train, ax=axes[0], fill=True)
axes[0].set_title('Distribution of Original Malic acid')

sns.kdeplot(x='Malic acid', data=X_train_s, ax=axes[1], fill = True)
axes[1].set_title('Distribution of Scaled Malic acid')

plt.show()
```
