### Data Pre-Processing

```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
movies.head()
```

```python
movies.shape
```

```python
movies.info()
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
credits.head()
```

```python
credits.shape
```

```python
credits.info()
```

```python
movies = movies.merge(credits, on='title')
movies.head()
```

```python
movies.info()
```

```python
# necessary features for our project:

# genres
# id
# keywords
# title
# overview
# cast
# crew
```

```python
movies = movies[['id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movies
```

```python
movies.info()
```

```python
movies.dropna(inplace=True)
movies.head()
```

```python
movies.info()
```

```python
movies['genres'][0]
```

```python
import ast # as the list is in string format we can see above
```

```python
def extract(content):
    the_list = []
    for i in ast.literal_eval(content):
        the_list.append(i['name'])
    return the_list
```

```python
extract(movies['genres'][0])
```

```python
movies['genres'] = movies['genres'].apply(extract)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(extract)
movies.head()
```

```python
movies['cast'][0]
```

```python
def extract_3(content):
    counter = 0
    the_list = []
    for i in ast.literal_eval(content):
        if counter==3:
            break
        else:
            the_list.append(i['name'])
        counter += 1
    return the_list
```

```python
extract_3(movies['cast'][0])
```

```python
movies['cast'] = movies['cast'].apply(extract_3)
movies.head()
```

```python
movies['crew'][0]
```

```python
def extract_director(content):
    the_list = []
    for i in ast.literal_eval(content):
        if i['job']=='Director':
            the_list.append(i['name'])
            break
    return the_list
```

```python
extract_director(movies['crew'][0])
```

```python
movies['crew'] = movies['crew'].apply(extract_director)
movies.head()
```

```python
def fill_space(content):
    the_list = []
    for i in content:
        the_list.append(i.replace(" ",""))
    return the_list
```

```python
fill_space(['Sam Worthington', 'Zoe Saldana', 'Sigourney Weaver'])
```

```python
fill_space(movies['genres'][0])
```

```python
movies['genres'] = movies['genres'].apply(fill_space)
movies['keywords'] = movies['keywords'].apply(fill_space)
movies['cast'] = movies['cast'].apply(fill_space)
movies['crew'] = movies['crew'].apply(fill_space)
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies['overview'][0]
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
movies = movies.drop(columns=['overview', 'genres', 'keywords', 'cast', 'crew'])
movies
```

```python
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))
movies
```

```python
movies['tags'] = movies['tags'].apply(lambda x:x.lower())
movies.head()
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
ps.stem('dances')
```

```python
def stemmed(text):
    the_list = []
    for i in text.split():
        the_list.append(ps.stem(i))
    return " ".join(the_list)
```

```python
stemmed(movies['tags'][0])
```

```python
stemmed(movies['tags'][0])
```

```python
movies['tags'] = movies['tags'].apply(stemmed)
```

```python
movies['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(movies['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
print(cv.get_feature_names_out())
```

```python
for i in cv.get_feature_names_out():
    print(i)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity.shape
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key = lambda x:x[1])[1:6]
```

```python
def recommend(movie_title):
    similarity_index = movies[movies['title'] == movie_title].index[0]
    recommended_list = sorted(list(enumerate(similarity[similarity_index])), reverse=True, key = lambda x:x[1])[1:6]

    top_5 = []
    for i in recommended_list:
        top_5.append(movies.iloc[i[0]].title)
    return top_5
```

```python
recommend('The Dark Knight')
```
